# **Used Cars Price Prediction**

### **Proble statement**

Build a regression model that predicts the prices of used cars based on available features (such as power of the engine, age, distance driven, transmission and fuel, locality, and brand).   


### **Solution summary**

- It is essential for a new company to set the price of the used cars according to the general trend. The regression model we built predicts the price of a car with 0.2 percent error, based on power of the engine, mileage, kilometers driven, manufacturing year, fuel_type, and transmission. The model accounts for 90% variation in the data.

- We propose to use the linear regression because it provides a good fit to data and high accuracy.  

- Using this model to predict prices of the used cars will increase company efficiency. It will also help in making good business decisions about buying and selling used cars to make profit.

### **Recommendations for implementation**

- This model should be used to predict car prices for the used cars, based on their characteristics. This will provide the starting point for setting the price for the cars the company has for sale.
- The larger data set has to be collected to include more important variables in the model, such as car brands and locations.  These features were not used in the final mofdel because there were not enough data to make them statistically significant.
- Some of the variables in the data set were highly correlated, for example power and engine capacity. It is recommended to expand this dataset with more features that are not dependent.
- Since the market is fluent, the model has to be periodically updated to incorporate new trends.  






### **Data Dictionary**

**S.No.** : Serial Number

**Name** : Name of the car which includes Brand name and Model name

**Location** : The location in which the car is being sold or is available for purchase (Cities)

**Year** : Manufacturing year of the car

**Kilometers_driven** : The total kilometers driven in the car by the previous owner(s) in KM

**Fuel_Type** : The type of fuel used by the car (Petrol, Diesel, Electric, CNG, LPG)

**Transmission** : The type of transmission used by the car (Automatic / Manual)

**Owner** : Type of ownership

**Mileage** : The standard mileage offered by the car company in kmpl or km/kg

**Engine** : The displacement volume of the engine in CC

**Power** : The maximum power of the engine in bhp

**Seats** : The number of seats in the car

**New_Price** : The price of a new car of the same model in INR 100,000

**Price** : The price of the used car in INR 100,000 (**Target Variable**)

```python
# Defining a function to plot a boxplot and a histogram with shared x-axis

def histogram_boxplot(data, feature, figsize = (12, 7), kde = False, bins = None):
    """
    Boxplot and histogram combined

    data: dataframe
    feature: dataframe column
    figsize: size of figure (default (12,7))
    kde: whether to show density curve (default False)
    bins: number of bins for histogram (default None)
    """
    f2, (ax_box2, ax_hist2) = plt.subplots(
        nrows = 2,      # Number of rows of the subplot grid = 2
        sharex = True,  # x-axis will be shared among all subplots
        gridspec_kw = {"height_ratios": (0.25, 0.75)},
        figsize = figsize,
    )                   # Creating the 2 subplots
    sns.boxplot(
        data = data, x = feature, ax = ax_box2, showmeans = True, color = "violet"
    )                   # Boxplot will be created and a star will indicate the mean value of the column
    sns.histplot(
        data = data, x = feature, kde = kde, ax = ax_hist2, bins = bins, palette = "blue"
    ) if bins else sns.histplot(
        data = data, x = feature, kde = kde, ax = ax_hist2
    )                   # For histogram
    ax_hist2.axvline(
        data[feature].mean(), color = "red", linestyle = "--"
    )                   # Add mean to the histogram
    ax_hist2.axvline(
        data[feature].median(), color = "orange", linestyle = "-"
    )                   # Add median to the histogram

    #ax_hist2.annotate((data[feature].mean()),('{:.0f}'.format(data[feature].mean()) , 0), size=10, ha='center')  # Annotate the mean with center alignment
```

### **Missing value treatment**

```python
# Making copy of the data frame, so that not to loose data.
data=df.copy()
```

```python
#Removing features that were replaced by new features
data.drop(['Fuel_Type','Seats','Brand'], axis=1, inplace=True)
```

```python
data.isnull().sum()
```

Mileage, Engine, Power, Seats_category have missing values. (Mileage also has zero values that have to be imputed.)
First, we will use median of the feature over the cars with the same Name.

```python
#Incerting median of the number of seats of the cars with the same Name
index_null=data[data['Seats_category'].isnull()].index
for i in index_null:
  a=data[data["Name"]==data.loc[i, 'Name']]['Seats_category'].median()
  data.loc[i, 'Seats_category']=a
```

```python
#Incerting median of 'Power' of the cars with the same Name
index_null=data[data['Power'].isnull()].index
for i in index_null:
  a=data[data["Name"]==data.loc[i, 'Name']]['Power'].median()
  data.loc[i, 'Power']=a
```

```python
#Incerting median of 'Engine' of the cars with the same Name
index_null=data[data['Engine'].isnull()].index
for i in index_null:
  a=data[data["Name"]==data.loc[i, 'Name']]['Engine'].median()
  data.loc[i, 'Engine']=a
```

```python
#Incerting median of 'Mileage' of the cars with the same Name. We treat zero entries as mising.
index_null=data[(data['Mileage'].isnull())|(data['Mileage']==0)].index
for i in index_null:
  a=data[data["Name"]==data.loc[i, 'Name']]['Mileage'].median()
  data.loc[i, 'Mileage']=a
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Function to check VIF
def checking_vif(train):
    vif = pd.DataFrame()
    vif["feature"] = train.columns

    # Calculating VIF for each feature
    vif["VIF"] = [
        variance_inflation_factor(train.values, i) for i in range(len(train.columns))
    ]
    return vif

print(checking_vif(X_train_scaled))
```

```python
# Function to perform likelihood ratio test
def ll_ratio_test(full_model, red_model, df):
  full_ll=full_model.llf
  red_ll=red_model.llf
  LR_statistic = 2*(full_ll-red_ll)
  p_val = scipy.stats.chi2.sf(LR_statistic, df)
  return p_val


```

#Performance metrics

Function to calculate performance metrics

```python
from sklearn.metrics import r2_score, mean_absolute_percentage_error, mean_absolute_error, mean_squared_error

# Model Performance on test and train data
def model_pref(olsmodel, x_train, x_test):

    # In-sample Prediction
    y_pred_train = olsmodel.predict(x_train)
    y_observed_train = y_train

    # Prediction on test data
    y_pred_test = olsmodel.predict(x_test)
    y_observed_test = y_test

    print(
        pd.DataFrame(
            {
                "Data": ["Train", "Test"],
                 "MSE": [
                  mean_squared_error(y_pred_train, y_observed_train),
                  mean_squared_error(y_pred_test, y_observed_test),],

                "RMSE": [
                    np.sqrt(mean_squared_error(y_pred_train, y_observed_train)),
                    np.sqrt(mean_squared_error(y_pred_test, y_observed_test)),
                ],
                "MAE": [
                    mean_absolute_error(y_pred_train, y_observed_train),
                    mean_absolute_error(y_pred_test, y_observed_test),
                ],
                "MAPE": [
                    mean_absolute_percentage_error(y_pred_train, y_observed_train),
                    mean_absolute_percentage_error(y_pred_test, y_observed_test),
                ],

                "r2": [
                    r2_score(y_pred_train, y_observed_train),
                    r2_score(y_pred_test, y_observed_test),
                ],
            }
        )
    )
```

```python
# Checking the ols_res_1 performance. (All the categorical features in)
model_pref(ols_res_1, X_train_scaled, X_test_scaled)
```

Performance on the training and test sets are similar. This indicates the model is not overfitted.
